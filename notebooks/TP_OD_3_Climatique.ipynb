```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from scipy import stats
import numpy as np
from datetime import datetime

# =============================================
# CONFIGURATION
# =============================================
DISASTER_PATH = r"C:\Users\nguet\Majeures\Open Data\TP_OD_3_Climatique (1)\Fusion_Disaster_1900_2021.csv"
OUTPUT_DIR = r"C:\Users\nguet\Majeures\Open Data\TP_OD_3_Climatique (1)"

# =============================================
# 1. CHARGEMENT DES DONNÉES DE CATASTROPHES
# =============================================
def load_disaster_data(file_path):
    try:
        print(f"\nChargement du fichier catastrophes: {file_path}")
        
        if not os.path.exists(file_path):
            print(f"ERREUR: Le fichier {file_path} n'existe pas")
            return None
        
        # Chargement avec détection automatique des colonnes de date
        df = pd.read_csv(file_path, encoding='latin1', nrows=5)
        date_cols = [col for col in df.columns if 'date' in col.lower() or 'time' in col.lower()]
        
        if date_cols:
            df = pd.read_csv(file_path, encoding='latin1', parse_dates=date_cols)
        else:
            df = pd.read_csv(file_path, encoding='latin1')
        
        print("\nChargement réussi. Aperçu des données :")
        print(df.head(3))
        return df

    except Exception as e:
        print(f"\nERREUR lors du chargement: {str(e)}")
        return None

# Chargement des données catastrophes
df_disaster = load_disaster_data(DISASTER_PATH)
if df_disaster is None:
    exit()

# =============================================
# 2. CRÉATION DE DONNÉES CLIMATIQUES SYNTHÉTIQUES
# =============================================
def generate_climate_data(years):
    """Génère des données climatiques factices pour les années concernées"""
    np.random.seed(42)
    
    climate_data = pd.DataFrame({
        'Year': years,
        'Temperature': np.random.normal(15, 2, len(years)) + 0.03 * (years - years.min()),
        'CO2': np.random.normal(350, 50, len(years)) + 1.5 * (years - years.min()),
        'SeaLevel': np.random.normal(0, 5, len(years)) + 0.1 * (years - years.min()),
        'Precipitation': np.random.normal(100, 20, len(years))
    })
    
    # Ajout d'une tendance climatique réaliste
    climate_data['Temperature'] += 0.02 * (climate_data['Year'] - climate_data['Year'].min())
    climate_data['CO2'] += 2 * (climate_data['Year'] - climate_data['Year'].min())
    
    return climate_data

# Création des variables temporelles pour les catastrophes
date_col = next((col for col in df_disaster.columns if 'date' in col.lower()), None)

if date_col and pd.api.types.is_datetime64_any_dtype(df_disaster[date_col]):
    df_disaster['Year'] = df_disaster[date_col].dt.year
    df_disaster['Month'] = df_disaster[date_col].dt.month
    df_disaster['Decade'] = (df_disaster['Year'] // 10) * 10
else:
    print("\nAttention: Colonne de date non trouvée, création d'une colonne Year factice")
    df_disaster['Year'] = np.random.randint(1900, 2022, size=len(df_disaster))
    df_disaster['Decade'] = (df_disaster['Year'] // 10) * 10

# Génération des données climatiques
years = range(df_disaster['Year'].min(), df_disaster['Year'].max() + 1)
df_climate = generate_climate_data(pd.Series(years))

# Fusion des données
df_combined = pd.merge(df_disaster, df_climate, on='Year', how='left')

# =============================================
# 3. ANALYSES ET VISUALISATIONS
# =============================================
generated_plots = []

# 1. Évolution des catastrophes et variables climatiques
plt.figure(figsize=(14, 10))

# Nombre de catastrophes par année
ax1 = plt.subplot(2, 1, 1)
disaster_count = df_combined['Year'].value_counts().sort_index()
disaster_count.plot(kind='line', color='red', ax=ax1)
ax1.set_title('Nombre de Catastrophes par Année')
ax1.set_ylabel('Nombre de catastrophes')

# Variables climatiques
ax2 = plt.subplot(2, 1, 2)
climate_vars = ['Temperature', 'CO2', 'SeaLevel']
for var in climate_vars:
    ax2.plot(df_climate['Year'], df_climate[var], label=var)
ax2.set_title('Évolution des Variables Climatiques')
ax2.set_ylabel('Valeurs normalisées')
ax2.legend()

plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, 'evolution_comparaison.png'))
generated_plots.append('evolution_comparaison.png')

# 2. Corrélations
plt.figure(figsize=(10, 8))
corr_data = df_combined.groupby('Year').agg({
    'Temperature': 'mean',
    'CO2': 'mean',
    'SeaLevel': 'mean',
    'Disaster Type': 'count'
}).rename(columns={'Disaster Type': 'DisasterCount'})

corr_matrix = corr_data.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Corrélations entre Variables Climatiques et Catastrophes')
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, 'correlations_climatiques.png'))
generated_plots.append('correlations_climatiques.png')

# 3. Analyse par type de catastrophe
if 'Disaster Type' in df_combined.columns:
    plt.figure(figsize=(12, 8))
    sns.boxplot(data=df_combined, x='Disaster Type', y='Temperature')
    plt.title('Température Moyenne par Type de Catastrophe')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.savefig(os.path.join(OUTPUT_DIR, 'temperature_par_type.png'))
    generated_plots.append('temperature_par_type.png')

# =============================================
# 4. SAUVEGARDE ET RAPPORT
# =============================================
# Sauvegarde des données enrichies
output_path = os.path.join(OUTPUT_DIR, 'disaster_climate_combined.csv')
df_combined.to_csv(output_path, index=False)

# Rapport final
print("\n=== RAPPORT D'EXÉCUTION ===")
print(f"Données catastrophes chargées : {len(df_disaster)} lignes")
print(f"Données climatiques générées : {len(df_climate)} années")
print(f"Données combinées sauvegardées : {output_path}")
print(f"Visualisations générées : {', '.join(generated_plots)}")

print("\nColonnes disponibles dans le dataset enrichi :")
print(df_combined.columns.tolist())

print("\nAperçu des données climatiques générées :")
print(df_combined[['Year', 'Temperature', 'CO2', 'SeaLevel']].drop_duplicates().head())
```

```python
import pandas as pd
import plotly.express as px
import os

# Configuration
INPUT_PATH = r"C:\Users\nguet\Majeures\Open Data\TP_OD_3_Climatique (1)\disaster_climate_combined.csv"
OUTPUT_DIR = r"C:\Users\nguet\Majeures\Open Data\TP_OD_3_Climatique (1)"

# 1. Chargement des données
print("Chargement des données...")
try:
    df = pd.read_csv(INPUT_PATH)
    print("Colonnes disponibles:", df.columns.tolist())
except Exception as e:
    print("Erreur de chargement:", str(e))
    exit()

# 2. Carte des températures corrigée
if all(col in df.columns for col in ['Latitude', 'Longitude', 'Temperature']):
    try:
        # Nettoyage des données
        temp_df = df.dropna(subset=['Latitude', 'Longitude', 'Temperature']).copy()
        
        # Conversion des types
        temp_df['Latitude'] = pd.to_numeric(temp_df['Latitude'], errors='coerce')
        temp_df['Longitude'] = pd.to_numeric(temp_df['Longitude'], errors='coerce')
        temp_df['Temperature'] = pd.to_numeric(temp_df['Temperature'], errors='coerce')
        temp_df = temp_df.dropna(subset=['Latitude', 'Longitude', 'Temperature'])
        
        # Création de la carte
        fig_temp = px.density_mapbox(
            temp_df,
            lat='Latitude',
            lon='Longitude',
            z='Temperature',
            radius=15,  # Ajustement du rayon
            center=dict(lat=20, lon=0),  # Centre de la carte
            zoom=1,  # Niveau de zoom initial
            mapbox_style="stamen-terrain",  # Style alternatif: "open-street-map"
            title="Distribution des températures anormales",
            hover_name='Country' if 'Country' in temp_df.columns else None,
            hover_data=['Year', 'Disaster Type'] if 'Year' in temp_df.columns and 'Disaster Type' in temp_df.columns else None,
            color_continuous_scale=px.colors.sequential.thermal,  # Échelle de couleurs adaptée
            range_color=[temp_df['Temperature'].min(), temp_df['Temperature'].max()]  # Plage de valeurs
        )
        
        # Ajustements supplémentaires
        fig_temp.update_layout(
            margin=dict(l=0, r=0, t=30, b=0),
            coloraxis_colorbar=dict(
                title="Température",
                thicknessmode="pixels",
                thickness=15,
                lenmode="pixels",
                len=300,
                yanchor="middle",
                y=0.5
            )
        )
        
        # Sauvegarde
        output_file = os.path.join(OUTPUT_DIR, "carte_temperatures_corrigee.html")
        fig_temp.write_html(output_file)
        print(f"\nCarte des températures sauvegardée sous: {output_file}")
        
    except Exception as e:
        print("\nErreur lors de la création de la carte des températures:", str(e))
else:
    missing_cols = [col for col in ['Latitude', 'Longitude', 'Temperature'] if col not in df.columns]
    print(f"\nImpossible de créer la carte des températures - colonnes manquantes: {missing_cols}")

print("\nTraitement terminé.")
```

```python
import pandas as pd
import plotly.express as px
import os

# Configuration
INPUT_PATH = r"C:\Users\nguet\Majeures\Open Data\TP_OD_3_Climatique (1)\disaster_climate_combined.csv"
OUTPUT_DIR = r"C:\Users\nguet\Majeures\Open Data\TP_OD_3_Climatique (1)"

# 1. Chargement des données
print("Chargement des données...")
try:
    df = pd.read_csv(INPUT_PATH)
    print("Colonnes disponibles:", df.columns.tolist())
except Exception as e:
    print("Erreur de chargement:", str(e))
    exit()

# 2. Vérification des colonnes nécessaires
required_columns = {'Country', 'Disaster Type'}
available_columns = set(df.columns)
missing_columns = required_columns - available_columns

if missing_columns:
    print(f"\nAttention: Colonnes manquantes {missing_columns}. Le script continue avec les colonnes disponibles.")

# 3. Carte simplifiée des catastrophes par pays
try:
    # Préparation des données
    country_stats = df.groupby('Country', as_index=False).agg(
        Nombre_catastrophes=('Disaster Type', 'count')
    )
    
    # Ajout des données climatiques si disponibles
    if 'Temperature' in df.columns:
        country_stats['Température_moyenne'] = df.groupby('Country')['Temperature'].mean().values
    
    # Création de la carte
    fig = px.choropleth(
        country_stats,
        locations="Country",
        locationmode="country names",
        color="Nombre_catastrophes",
        hover_name="Country",
        hover_data=["Température_moyenne"] if 'Température_moyenne' in country_stats.columns else None,
        title="Nombre de catastrophes par pays",
        color_continuous_scale=px.colors.sequential.Plasma
    )
    
    # Sauvegarde
    output_file = os.path.join(OUTPUT_DIR, "carte_catastrophes_pays.html")
    fig.write_html(output_file)
    print(f"\nCarte sauvegardée sous: {output_file}")

except Exception as e:
    print("\nErreur lors de la création de la carte:", str(e))

# 5. Carte des risques par type de catastrophe
if 'Disaster Type' in df.columns:
    try:
        disaster_counts = df['Disaster Type'].value_counts().reset_index()
        disaster_counts.columns = ['Type', 'Count']
        
        fig_pie = px.pie(
            disaster_counts,
            names='Type',
            values='Count',
            title="Répartition des types de catastrophes"
        )
        output_file = os.path.join(OUTPUT_DIR, "repartition_catastrophes.html")
        fig_pie.write_html(output_file)
        print(f"Graphique de répartition sauvegardé sous: {output_file}")
    except Exception as e:
        print("Erreur création graphique de répartition:", str(e))

print("\nTraitement terminé. Vérifiez le dossier de sortie pour les résultats.")
```
